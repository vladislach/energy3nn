```python
import numpy as np
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt


hyperparams = {'LMAX': [0, 1, 2, 3, 4],
               'NUM_BASIS': [6, 8, 10],
               'BNI': [8, 16, 32, 64],
               'INVARIANT_LAYERS': [1, 2, 3],
               'INVARIANT_NEURONS': [16, 32, 64, 128],
               'NUM_CONV_LAYERS': [1, 2, 3, 4, 5],
               'RESNET': [True, False],
               'USE_SC': [True, False],
               'BATCH_SIZE': [2, 5, 10],
               'LEARNING_RATE': [1e-2, 5e-3, 1e-3, 1e-4, 1e-5]}

for hp_name in hyperparams:
    num_cols = len(hyperparams[hp_name])
    fig, axes = plt.subplots(1, num_cols, figsize=(num_cols * 5, 4))

    for i, hp_val in enumerate(hyperparams[hp_name]):
        
        dir = f"outputs/model_{hp_name}_{hp_val}"
        train_losses = np.load(dir + '/train_losses.npy')
        val_losses = np.load(dir + '/test_losses.npy')
        axes[i].plot(train_losses, label="Train Loss", linewidth=1.5, color='royalblue')
        axes[i].plot(val_losses, label="Test Loss", linewidth=1.5, color='lightseagreen')
        axes[i].set_title(f"{hp_name}={hp_val}")
        axes[i].legend()

    plt.show()
```
